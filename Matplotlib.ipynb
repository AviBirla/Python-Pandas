# **A picture is worth a thousand words**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[30,32,35,32,34,30,29]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(x,y,color='red',linewidth=4,linestyle='dashdot')
```

```python
plt.plot(x,y,color='blue',marker="o",linestyle="--",linewidth=2,markersize=6)      # can also use hexa color code in sting and alpha ( range is b/w 0 to 1 ....for tranperancy)
```

### Axis labels , Legend & Grid :

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
plt.xlabel("Days")
plt.ylabel("Temperatur")
plt.title("weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(shadow=True,fontsize='large')     # plt.legend(loc='best')----. bydefault location ,,,, loc ; upper/lower right, upper/lower left ,center, lower/upper center,etc.
plt.grid()
```

# Histograms

```python
blood_sugar_men=[113,85,90,150,149,88,95,115,135,80,77,82,129]
blood_sugar_women=[67,95,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('Sugar range')
plt.ylabel('Total no. of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men , blood_sugar_women],bins=[80,100,125,160],rwidth=0.90,color=['g','orange'],label=['men','women'])       # histtype='step' , orientation='horizontal'
plt.legend()                                                   
```
