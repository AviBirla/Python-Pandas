# **Handling Missing Data : fillna , interpolate & dropna**

```python
new_df=df.fillna(0)    
new_df
```

```python
#sometime replacing NaN with 0 is not probably the best guess !!bcoz in column of event ... 0 seems irrelevant.
new_df=df.fillna({
    
    'temperature' : 0,
    'windspeed':0,
    'event': 'No event'
})
new_df
```

```python
new_df=df.fillna(method='ffill')      #forward fill the previous values in place of NaN ...this looks little better than having just 0 value
new_df                                # you can also use 'bfill' i.e backward fill 
```

```python
new_df=df.fillna(method='ffill',limit=1)      # limit =int, default None If method is specified, this is the maximum number of consecutive NaN values to forward/backward fill. In other words, if there is a gap with more than this number of consecutive NaNs, it will only be partially filled. If method is not specified, this is the maximum number of entries along the entire axis where NaNs will be filled. Must be greater than 0 if not None.
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
# if you want to drop all  complete rows containing NaN :
new_df=df.dropna() 
new_df
```

```python
# droping only those rows whole all values are NaN
new_df=df.dropna(how='all') 
new_df
```

```python
#.keep that row which have atleat one non-NaN/valid value .
new_df=df.dropna(thresh=1)         
new_df
#thresh=2 ... atleast 2 valid values
```

```python
#if you want to add missing values like we dont have 2nd,3rd,9th January in our data :
dt=pd.date_range('01-01-2017','01-11-2017')
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
